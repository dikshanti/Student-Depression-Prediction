## Data Preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
categorical_cols = ['Gender', 'Sleep Duration', 'Dietary Habits', 
                    'Have you ever had suicidal thoughts ?', 'Family History of Mental Illness','Depression']

label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    Students[col] = le.fit_transform(Students[col])
    label_encoders[col] = le
```

```python
Student_corr = Students.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(Student_corr, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Student Data')
plt.tight_layout()
plt.show()
```

```python
X = Students.drop(columns=['Depression'])
y = Students['Depression']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("\n Classification Report:\n", classification_report(y_test, y_pred))
print("\n Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Label encoders for categorical data
gender_mapping = {'Male': 0, 'Female': 1}
dietary_mapping = {'Healthy': 0, 'Moderate': 1, 'Unhealthy': 2}
yes_no_mapping = {'Yes': 1, 'No': 0}
sleep_mapping = {'Less than 5 hours': 0, '5-6 hours': 1, '7-8 hours': 2, 'More than 8 hours': 3}

# Function to take user input and make predictions
def predict_depression():
    print("=== Depression Prediction ===")
    print("Please provide the following details:\n")
    
    # Take user inputs
    gender = input("Gender (Male/Female): ")
    while gender not in gender_mapping:
        print("Invalid input. Please enter Male or Female.")
        gender = input("Gender (Male/Female): ")
    
    age = int(input("Age (e.g., 25): "))
    
    academic_pressure = float(input("Academic Pressure (1-5, e.g., 3.0): "))
    study_satisfaction = float(input("Study Satisfaction (1-5, e.g., 4.0): "))
    
    sleep_duration = input("Sleep Duration (Less than 5 hours/5-6 hours/7-8 hours/More than 8 hours): ")
    while sleep_duration not in sleep_mapping:
        print("Invalid input. Choose from: Less than 5 hours, 5-6 hours, 7-8 hours, More than 8 hours.")
        sleep_duration = input("Sleep Duration (Less than 5 hours/5-6 hours/7-8 hours/More than 8 hours): ")
    
    dietary_habits = input("Dietary Habits (Healthy/Moderate/Unhealthy): ")
    while dietary_habits not in dietary_mapping:
        print("Invalid input. Choose from: Healthy, Moderate, Unhealthy.")
        dietary_habits = input("Dietary Habits (Healthy/Moderate/Unhealthy): ")
    
    suicidal_thoughts = input("Have you ever had suicidal thoughts? (Yes/No): ")
    while suicidal_thoughts not in yes_no_mapping:
        print("Invalid input. Please enter Yes or No.")
        suicidal_thoughts = input("Have you ever had suicidal thoughts? (Yes/No): ")
    
    study_hours = int(input("Study Hours (e.g., 8): "))
    financial_stress = int(input("Financial Stress (1-5, e.g., 3): "))
    
    family_history = input("Family History of Mental Illness (Yes/No): ")
    while family_history not in yes_no_mapping:
        print("Invalid input. Please enter Yes or No.")
        family_history = input("Family History of Mental Illness (Yes/No): ")
    
    # Convert inputs to encoded values
    gender = gender_mapping[gender]
    sleep_duration = sleep_mapping[sleep_duration]
    dietary_habits = dietary_mapping[dietary_habits]
    suicidal_thoughts = yes_no_mapping[suicidal_thoughts]
    family_history = yes_no_mapping[family_history]
    
    # Prepare input for the model
    input_features = np.array([[gender, age, academic_pressure, study_satisfaction,
                                 sleep_duration, dietary_habits, suicidal_thoughts,
                                 study_hours, financial_stress, family_history]])
    
    # Make a prediction
    prediction = model.predict(input_features)
    
    # Display the result
    result = "Depressed" if prediction[0] == 1 else "Not Depressed"
    print("\n=== Prediction Result ===")
    print(f"The prediction is: {result}")

# Run the function to predict depression
predict_depression()
```
